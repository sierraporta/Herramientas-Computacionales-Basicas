```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
def f(L):
    return 2*np.pi*np.sqrt(L/980)

Lvals = np.arange(10., 110, 10)
Tvals = f(Lvals)
```

```python
Lvals
```

```python
Tvals
```

```python
# Generate data with a bit of randomness
# (the noise-less function that underlies the data is shown as a blue line)

Ldata = np.array(Lvals)
np.random.seed(42)
err_stdev = 0.2
Tvals_err =  np.random.normal(0., err_stdev, len(Ldata))
Tdata = f(Ldata) + Tvals_err
```

```python
Ldata
```

```python
Tdata
```

```python
Tdata
```

```python
plt.figure(figsize=(8,6))
plt.plot(Lvals,Tvals)
plt.scatter(Ldata,Tdata)
plt.xlabel("$L$, (cm)",fontsize=16)
plt.ylabel("$T$, (s)",fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.plot(Lvals**(0.5),Tvals)
plt.scatter(Ldata**(0.5),Tdata)
plt.xlabel("$L$, (cm)",fontsize=16)
plt.ylabel("$T^{-0.5}$, (s)",fontsize=16)
plt.show()
```

```python
np.corrcoef(Ldata**(0.5),Tdata)
```

```python
from scipy.optimize import curve_fit

def func1(x, a, b):
    return a*x**(1./2)+b

popt1, pcov1 = curve_fit(func1, Ldata, Tdata)
```
